```python
import pandas as pd
import random
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
# 1. Gaze Data import and computation of X and Y gaze coordinates
```

```python
#Read gaze data as using pandas library
gazeData = pd.read_csv("inputData/gazeData.csv", skiprows=25)
# set display.max_columns to none, to show all the columns when using gazeData.head() to preview the gazeData
pd.set_option('display.max_columns', None)
```

```python
# preview gaze data
# Note: see inputData/gazeData.csv for columns meta-data (under Section #METADATA)
gazeData.head()
```

```python
# compute the mean X and Y positions of the gazes based on the X-coordinate of the gaze points of the left and right eyes
gazeData["GazeX"] = gazeData[["ET_GazeLeftx","ET_GazeRightx"]].mean(axis=1)
gazeData["GazeY"] = gazeData[["ET_GazeLefty","ET_GazeRighty"]].mean(axis=1)
```

```python
# preview gaze data with mean X and Y gaze positions columns
gazeData.head()
```

```python
# 2. Plot gaze data and AOIs; Assign gaze points to AOIs
```

```python
# Read file with AOI coordinates
aois = pd.read_csv("inputData/aois_definition.csv")
```

```python
# Display the user defined areas of interest
aois
```

```python
# Plot gazes and AOIs


def initPlotContainer(backgroundImagePath):
    """ Method Description: iniate a coordinate system with a background image
        Method Arguments: 
                backgroundImagePath: path of the background image
         Method Return: fig (figure), ax (array of axes) """   
    # set plot dimensions
    plt.rcParams['figure.figsize'] = [40, 20]
    # initiate plot
    fig, ax = plt.subplots()
    #invert axses
    ax.invert_yaxis()
    # plot stimulus as background
    img = plt.imread(backgroundImagePath)
    ax.imshow(img)
    return fig, ax

def generateRandomColors(nColors):
    # generate a list of random colors
    colors = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
                 for i in range(nColors)]
    return colors

fig, ax = initPlotContainer("stimulus.png")
# plot gazes
ax.plot(gazeData["GazeX"], gazeData["GazeY"],'o',alpha=0.5, markersize=1, color='red')
# generate a list of random colors (to be assigned to the different AOIs)
colors = generateRandomColors(12)
# plot AOIs
# iterate over the rows in aois gazeDataframe
for index, row in aois.iterrows():
    # plot each AOI as a rectangle
    ax.add_patch(Rectangle((row["xStart"], row["yStart"]), row["width"], row["height"],color=colors[index],alpha=0.2))
    # add text with the AOI name     
    ax.annotate(row["AOI"], (row["xStart"]+row["width"]/2, row["yStart"]+row["height"]/2), color='blue', weight='bold', fontsize=25, ha='center', va='center')
```

```python
# Assign gaze points to AOIs
def pointInRect(px,py,x1,y1,w,h):
    """ Method Description: check if a point is within a rectrangle
        Method Arguments: 
                px: point x cooridinate value
                py: point y cooridinate value
                x1: retangle starting point in x coordinate 
                y1: retangle starting point om y coordinate 
                w: retangle width
                h: rectangle height
         Method Return: 0 or 1 (0: gaze point is not in AOI, 1: gaze point is within AOI)"""

    x2 = x1+w
    y2 = y1+h
    return 1 if px >= x1 and px <= x2 and py >= y1 and py <= y2 else 0; 

# copy gazeData to gazeDataWithAOIs
gazeDataWithAOIs = gazeData.copy(deep=True)
# iterate over the rows in aois gazeDataframe
for index, row in aois.iterrows():
    # get the name of AOI
    aoiName = row["AOI"]
    # add a new column with key=aoiName and value=(0 or 1) depending on the pointInRect return value. 
    gazeDataWithAOIs[aoiName] = gazeData.apply(lambda x : pointInRect(x["GazeX"],x["GazeY"],row["xStart"],row["yStart"],row["width"],row["height"]),axis=1)
```

```python
# preview gaze data with AOIs mapping
gazeDataWithAOIs.head()
```

```python
# Save gazeData with AOIs
gazeDataWithAOIs.to_csv("outData/gazeDataWithAOIs.csv")
```
