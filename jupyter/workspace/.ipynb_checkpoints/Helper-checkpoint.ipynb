```python
import os
import json
import csv

def parse_coords(coord_string):
    x_str, y_str = coord_string.split(',')
    return float(x_str), float(y_str)

def transform_point_adjusted(point, translation, scale):
    # Apply scaling
    scaled_point = (point[0] * scale[0], point[1] * scale[1])
    
    # Apply translation
    translated_point = (scaled_point[0] + translation[0], scaled_point[1] + translation[1])
    
    return scaled_point

def calculate_opposite_corners_adjusted(p1, p2):
    # Calculate opposite corners
    return [(p1[0], p2[1]), (p2[0], p1[1])]

# ... (Rest of the script remains the same)

# The main block where transformation is applied:



# Folder containing JSON files
folder_path = 'test/aois'

# CSV file to store the results
csv_file_path = 'test/aois.csv'

# Iterate over JSON files and process them
with open(csv_file_path, mode='w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['File', 'Point1', 'Point2', 'Point3', 'Point4'])

    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            file_path = os.path.join(folder_path, filename)

            with open(file_path, 'r', encoding='UTF-16') as json_file:
                data = json.load(json_file)

                point1 = parse_coords(data["GeometryData"]["Points"][0])
                point2 = parse_coords(data["GeometryData"]["Points"][1])
                translation = parse_coords(data["GeometryData"]["Transform"]["Translation"])
                scale = parse_coords(data["GeometryData"]["Transform"]["Scale"])

                transformed_point1 = transform_point_adjusted(point1, translation, scale)
                transformed_point2 = transform_point_adjusted(point2, translation, scale)
                opposite_corners = calculate_opposite_corners_adjusted(transformed_point1, transformed_point2)
                final_coordinates = [transformed_point1, transformed_point2] + opposite_corners

                writer.writerow([filename] + final_coordinates)

import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import csv
import random

# Path to the image file
image_path = 'test/stimulus.png'
# Path to the CSV file
csv_file_path = 'test/aois.csv'

# Open the image
img = Image.open(image_path)
fig, ax = plt.subplots(figsize=(20, 11))
ax.imshow(img)

# Function to generate random colors
def random_color():
    return [random.random() for _ in range(3)]

# Read the coordinates from the CSV file and plot the rectangles
with open(csv_file_path, mode='r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        # Extract coordinates for each rectangle
        points = [tuple(map(float, point.strip('()').split(','))) for point in row[1:]]
        # Create a rectangle patch
        rect = patches.Polygon(points, closed=True, edgecolor=random_color(), fill=False)
        # Add the patch to the Axes
        ax.add_patch(rect)

# Display the plot
plt.show()

```
